```python
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
import pandas
import numpy as np

transaction_H_filter_10 = pandas.read_csv('transaction_H_filter_10.csv')

print(transaction_H_filter_10)

```

```python
from random import sample
import math

all_rules_H = pandas.DataFrame()
for i in range(100):
    sub = sample(set(transaction_H_filter_10.index),math.floor(len(transaction_H_filter_10.index)*0.75))
    trans_sub = transaction_H_filter_10.loc[sub]
    frq_items = fpgrowth(trans_sub, min_support = 0.7, use_colnames = True)
    for j in frq_items.index:
        if (abs(transaction_H_filter_10[frq_items.itemsets[j]].corr())<0.1).any().any():
            frq_items = frq_items.drop(j)
    rules = association_rules(frq_items, metric ='lift', min_threshold = 1.01)
    all_rules_H = pandas.concat([all_rules_H, rules])
    
    
print(all_rules_H)
```

```python
all_rules_H.to_csv("python_rules_H_0.1pre_0.7supp.csv", sep='\t', encoding='utf-8')
```
